```python
%matplotlib inline
from matplotlib import pyplot as plt
from pylab import *

t = arange(0.2,2.0,0.01)
s = sin(2*pi*t)
c = cos(2*pi*t)

plt.plot(t,s, color="green", linewidth=2, alpha=0.5, label="sin(x)")
plt.plot(t,c, color="orange", linewidth=2, alpha=0.5, label="cos(x)")
plt.legend(loc="upper left")
```

```python
import numpy as np
mean, cov = [0,1],[(1,.5),(.5,1)]
data = np.random.multivariate_normal(mean,cov,500)
plt.plot(data[:,0],data[:,1], marker='o', linestyle=' ')
plt.xlabel("My x-variable")
plt.ylabel("My y-variable")
savefig("my_png.png")
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame(data, columns=['x','y'])
sns.jointplot(x="x", y="y", data=df, kind="kde")
```
